```python
#Exercise 1
class shoppingbag():
    def __init__(self, handles, capacity, items):
        self.handles =  handles
        self.capacity = capacity
        self.items = items


    def showitems(self):
        print(f'you have items in your bag')
        for item in self.items:
            print(item)

    def showcapacity(self):
        print(f"your capacity is {self.capacity} ")

    def additems(self):
        products = input('what would you like to add? ')
        self.items.append(products)

    def changebagcap(self, capacity):
        self.capacity = capacity

    def increasecap(self, changed_capacity = 10):
        if self.capacity == isinstance(self.capacity, str):
            print('invalid')
        else:
            self.capacity += changed_capacity

    def removeitems(self):
        product = input('What do you want to remove')
        self.items.remove(product)


wholefoods = shoppingbag(2 , 10, [ ])


def run():
    while True:
        response = input('What do you want to do? add or show cart or remove item')
        if response.lower() == "quit":
            wholefoods.showitems()
            print(f"thanks for shopping")
            break
        elif response.lower() == "add":
            wholefoods.additems()
        elif response.lower() == "show":
            wholefoods.showitems()
        elif response.lower() == 'remove':
            wholefoods.removeitems()
        else:
            print("try another command")


run()




# Exercise 2

class Car:
    def __init__(self, color, make, model, wheels):
        self.color = color
        self.make = make
        self.model = model
        self.wheels = wheels

    def iscar(self):
        print('This is a car')



class Ford(Car):
    def __init__(self, color, make, model, wheels):
        super(Ford, self).__init__(color, make, model, wheels)

    def iscar(self):
        print(f'This car is a {self.color} {self.make} {self.model} with {self.wheels} wheels')


vehicle = Ford('blue', 'ford', "explorer",4)
print(vehicle.iscar())
```
